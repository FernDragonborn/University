```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('city_happiness_train.csv')
print(f'Розмірність датасету: {df.shape}')
print(f'\nСтовпці: {list(df.columns)}')
print(f'\nТипи даних:\n{df.dtypes}')
```

```python
print(df.head(5))
print(df.head(5).T.to_string())
```

```python
df.info()
print('-'* 80)
print(df.T.describe())
```

```python
aqi_by_city = df.groupby('City')['Air_Quality_Index'].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
plt.bar(range(20), aqi_by_city.values[:20], color='steelblue')
plt.xticks(range(20), aqi_by_city.index[:20], rotation=45, ha='right')
plt.xlabel('Місто')
plt.ylabel('Air Quality Index (середнє)')
plt.title('Топ-20 міст за середнім індексом якості повітря (по спаданню)')
plt.tight_layout()
plt.savefig('aqi_bar.png', dpi=150)
plt.show()
```

```python
max_hi = df['Healthcare_Index'].max()
max_city = df[df['Healthcare_Index'] == max_hi]['City'].iloc[0]
print(f'Максимальний Healthcare Index: {max_hi}')
print(f'Місто: {max_city}')
```

```python
import seaborn as sns

df_cleaned = df[df['Happiness_Score'] >= 0].copy()
df_cleaned['Traffic_Density'] = df_cleaned['Traffic_Density'].astype('category')
plt.figure(figsize=(12, 7), dpi=100)

sns.scatterplot(
    data=df_cleaned,
    x='Air_Quality_Index',
    y='Happiness_Score',
    hue='Traffic_Density',
    alpha=0.5,
    s=60,
)
plt.title('Correlation: Air Quality vs Happiness Score', fontsize=15, pad=15)
plt.xlabel('Air Quality Index (AQI)', fontsize=12)
plt.ylabel('Happiness Score', fontsize=12)

plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```
