```python
import pygame
import numpy as np
from pygame.locals import *

K = 4
d = [
    np.random.uniform(-5.0, 18.0, 250),
    np.random.uniform(-8.0, 10.0, 250),
    np.random.uniform(0.0, 20.0, 250),
    np.random.uniform(-5.0, 36.0, 250)
]
eps = 0.5
max_iter = 200

q_a = [0.0] * K     # оцінки цінності дій
n_a = [0] * K       # кількість вибору кожної дії
score = 0
i = 0
```

```python
pygame.init()
WIDTH, HEIGHT = 800, 500
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption('K-Armed Bandit (ε-greedy)')
font = pygame.font.SysFont('Arial', 18)
big_font = pygame.font.SysFont('Arial', 24, bold=True)

COLORS = [(70, 130, 180), (220, 80, 80), (80, 180, 80), (200, 160, 50)]
BG_COLOR = (30, 30, 30)
TEXT_COLOR = (220, 220, 220)
```

```python
def draw_bar(surface, x, y, w, h, color, label, value):
    """Малює стовпчик з підписом."""
    bar_h = max(0, int(h * value / 20))
    pygame.draw.rect(surface, color, (x, y + h - bar_h, w, bar_h))
    pygame.draw.rect(surface, TEXT_COLOR, (x, y, w, h), 1)
    lbl = font.render(label, True, TEXT_COLOR)
    surface.blit(lbl, (x + w // 2 - lbl.get_width() // 2, y + h + 5))
    val = font.render(f'{value:.2f}', True, TEXT_COLOR)
    surface.blit(val, (x + w // 2 - val.get_width() // 2, y - 20))

def draw_info(surface, iteration, score, action_type, chosen):
    texts = [
        f'Iteration: {iteration}/{max_iter}',
        f'Total Score: {score:.2f}',
        f'Action: {action_type} (arm {chosen})',
        f'ε = {eps}'
    ]
    for j, t in enumerate(texts):
        txt = font.render(t, True, TEXT_COLOR)
        surface.blit(txt, (550, 30 + j * 25))
```

```python
running = True
clock = pygame.time.Clock()

while running:
    for event in pygame.event.get():
        if event.type == QUIT:
            running = False

    if i < max_iter:
        if np.random.random() < eps:
            action : int = np.random.randint(0, K)
            action_type = 'exploration'
        else:
            action = np.argmax(q_a)
            action_type = 'exploitation'

        reward = d[action][i % len(d[action])]
        score += reward

        n_a[action] += 1
        q_a[action] += (reward - q_a[action]) / n_a[action]

        i += 1
    else:
        action_type = 'done'
        action = -1

    screen.fill(BG_COLOR)

    title = big_font.render('K-Armed Bandit (ε-greedy)', True, TEXT_COLOR)
    screen.blit(title, (20, 10))

    bar_w = 80
    bar_h = 300
    for k in range(K):
        x = 50 + k * (bar_w + 30)
        draw_bar(screen, x, 80, bar_w, bar_h, COLORS[k],
                 f'Arm {k}', q_a[k])
        cnt = font.render(f'n={n_a[k]}', True, TEXT_COLOR)
        screen.blit(cnt, (x + bar_w // 2 - cnt.get_width() // 2,
                          80 + bar_h + 25))

    draw_info(screen, i, score, action_type, action)

    pygame.display.flip()
    clock.tick(30)

pygame.quit()
```

```python
import numpy as np
import pygame
from pygame.locals import *

GRID_SIZE = 5
gamma = 0.9
threshold = 0.001

A_pos = (1, 2)
A_prime = (3, 1)
A_reward = 7

actions = [(-1, 0), (1, 0), (0, -1), (0, 1)]
action_names = ['↑', '↓', '←', '→']
```

```python
V = np.zeros((GRID_SIZE, GRID_SIZE))

iteration = 0
while True:
    delta = 0
    V_new = np.copy(V)

    for r in range(GRID_SIZE):
        for c in range(GRID_SIZE):
            # Спеціальна клітинка A
            if (r, c) == A_pos:
                V_new[r, c] = A_reward + gamma * V[A_prime]
            else:
                # Звичайна клітинка — max по діях
                best_value = float('-inf')
                for dr, dc in actions:
                    nr, nc = r + dr, c + dc
                    if 0 <= nr < GRID_SIZE and 0 <= nc < GRID_SIZE:
                        # Всередині поля: reward = 0
                        value = 0 + gamma * V[nr, nc]
                    else:
                        # За межами: reward = -1, залишаємось
                        value = -1 + gamma * V[r, c]
                    best_value = max(best_value, value)

                V_new[r, c] = best_value

            delta = max(delta, abs(V_new[r, c] - V[r, c]))

    V = V_new
    iteration += 1

    if delta < threshold:
        break

print(f'Збіглося за {iteration} ітерацій')
print('Таблиця цінності станів v*:')
print(np.round(V, 2))
```

```python
policy = np.empty((GRID_SIZE, GRID_SIZE), dtype='U2')

for r in range(GRID_SIZE):
    for c in range(GRID_SIZE):
        if (r, c) == A_pos:
            policy[r, c] = '★'
            continue

        best_action = 0
        best_value = float('-inf')
        for a_idx, (dr, dc) in enumerate(actions):
            nr, nc = r + dr, c + dc
            if 0 <= nr < GRID_SIZE and 0 <= nc < GRID_SIZE:
                value = 0 + gamma * V[nr, nc]
            else:
                value = -1 + gamma * V[r, c]

            if value > best_value:
                best_value = value
                best_action = a_idx

        policy[r, c] = action_names[best_action]

print('Оптимальна політика π*:')
print(policy)
```

```python
pygame.init()
CELL = 80
MARGIN = 40
W = GRID_SIZE * CELL + 2 * MARGIN
H = GRID_SIZE * CELL + 2 * MARGIN + 60
screen = pygame.display.set_mode((W, H))
pygame.display.set_caption('Gridland — Value Iteration')

font = pygame.font.SysFont('Arial', 16)
big_font = pygame.font.SysFont('Arial', 22, bold=True)
arrow_font = pygame.font.SysFont('Arial', 28)

BG = (30, 30, 30)
GRID_COLOR = (100, 100, 100)
CELL_COLOR = (50, 50, 60)
A_COLOR = (70, 150, 70)
AP_COLOR = (150, 70, 70)

running = True
while running:
    for event in pygame.event.get():
        if event.type == QUIT:
            running = False

    screen.fill(BG)
    title = big_font.render(f'Gridland v* (γ={gamma}, iter={iteration})',
                             True, (220, 220, 220))
    screen.blit(title, (MARGIN, 10))

    for r in range(GRID_SIZE):
        for c in range(GRID_SIZE):
            x = MARGIN + c * CELL
            y = MARGIN + 40 + r * CELL

            # Колір клітинки
            if (r, c) == A_pos:
                color = A_COLOR
            elif (r, c) == A_prime:
                color = AP_COLOR
            else:
                # Градієнт за цінністю
                intensity = int(50 + 150 * (V[r, c] - V.min()) /
                                (V.max() - V.min() + 1e-9))
                color = (50, 50, intensity)

            pygame.draw.rect(screen, color, (x, y, CELL, CELL))
            pygame.draw.rect(screen, GRID_COLOR, (x, y, CELL, CELL), 1)

            # Значення цінності
            val_text = font.render(f'{V[r, c]:.1f}', True, (200, 200, 200))
            screen.blit(val_text, (x + 5, y + 5))

            # Стрілка політики
            arr = arrow_font.render(policy[r, c], True, (255, 255, 100))
            screen.blit(arr, (x + CELL//2 - arr.get_width()//2,
                              y + CELL//2 - arr.get_height()//2 + 5))

            # Підписи A, A'
            if (r, c) == A_pos:
                lbl = font.render('A (+7)', True, (255, 255, 255))
                screen.blit(lbl, (x + 5, y + CELL - 20))
            elif (r, c) == A_prime:
                lbl = font.render("A'", True, (255, 255, 255))
                screen.blit(lbl, (x + 5, y + CELL - 20))

    pygame.display.flip()

pygame.quit()
```
