```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl

economy = ctrl.Antecedent(np.arange(0, 2.1, 0.1), 'economy')
social = ctrl.Antecedent(np.arange(0, 2.1, 0.1), 'social_support')
happiness = ctrl.Consequent(np.arange(0, 10.1, 0.1), 'happiness_index')

economy.automf(3)
social.automf(3)

happiness['low'] = fuzz.trimf(happiness.universe, [0, 0, 5])
happiness['medium'] = fuzz.trimf(happiness.universe, [0, 5, 10])
happiness['high'] = fuzz.trimf(happiness.universe, [5, 10, 10])
```

```python
rule1 = ctrl.Rule(economy['poor'] & social['poor'], happiness['low'])
rule2 = ctrl.Rule(economy['poor'] & social['average'], happiness['low'])
rule3 = ctrl.Rule(economy['poor'] & social['good'], happiness['medium'])
rule4 = ctrl.Rule(economy['average'] & social['poor'], happiness['low'])
rule5 = ctrl.Rule(economy['average'] & social['average'], happiness['medium'])
rule6 = ctrl.Rule(economy['average'] & social['good'], happiness['high'])
rule7 = ctrl.Rule(economy['good'] & social['poor'], happiness['medium'])
rule8 = ctrl.Rule(economy['good'] & social['average'], happiness['high'])
rule9 = ctrl.Rule(economy['good'] & social['good'], happiness['high'])

happiness_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6, rule7, rule8, rule9])
happiness_sim = ctrl.ControlSystemSimulation(happiness_ctrl)
```

```python
from matplotlib import pyplot as plt
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

path = "world_happiness_report_2005_2025.csv"
df = pd.read_csv(path)
df.columns = df.columns.str.strip()

# Вибір 3 країн: poor, average, good за economy
df_2024 = df[df['year'] == 2024].copy()
df_2024 = df_2024.sort_values('explained_log_gdp_per_capita')

poor_country = df_2024.iloc[0]
avg_country = df_2024.iloc[len(df_2024)//2]
good_country = df_2024.iloc[-1]

test_countries = [poor_country, avg_country, good_country]

# нормалізація до діапазону universe (0-2)
scaler = MinMaxScaler(feature_range=(0, 2))
features = ['explained_log_gdp_per_capita', 'explained_social_support']
scaler.fit(df[features])

print("Тестування нечіткої системи на реальних даних (2024 рік):")
print("=" * 60)

for row in test_countries:
    raw_eco = row['explained_log_gdp_per_capita']
    raw_soc = row['explained_social_support']
    scaled = scaler.transform([[raw_eco, raw_soc]])[0]

    happiness_sim.input['economy'] = scaled[0]
    happiness_sim.input['social_support'] = scaled[1]

    happiness_sim.compute()
    fuzzy_score = happiness_sim.output['happiness_index']
    real_score = row['happiness_score']

    print(f"Країна: {row['country']}")
    print(f"  Економіка: {raw_eco:.2f} (scaled: {scaled[0]:.2f}), Соц. підтримка: {raw_soc:.2f} (scaled: {scaled[1]:.2f})")
    print(f"  Нечіткий прогноз: {fuzzy_score:.2f}")
    print(f"  Реальний індекс:  {real_score:.2f}")
    print(f"  Різниця: {abs(fuzzy_score - real_score):.2f}")
    print("-" * 60)

    economy.view(sim=happiness_sim)
    plt.title(f"Economy - {row['country']}")
    plt.show()
    social.view(sim=happiness_sim)
    plt.title(f"Social - {row['country']}")
    plt.show()
    happiness.view(sim=happiness_sim)
    plt.title(f"Happiness - {row['country']}")
    plt.show()
```
