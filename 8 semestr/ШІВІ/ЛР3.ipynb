```python
from typing import cast
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

path = "world_happiness_report_2005_2025.csv"
df: pd.DataFrame = cast(pd.DataFrame, pd.read_csv(path))
df.columns = df.columns.str.strip()
feature_cols = [
    'year',
    'explained_log_gdp_per_capita',
    'explained_social_support',
    'explained_healthy_life_expectancy',
    'explained_freedom',
    'explained_generosity',
    'explained_corruption'
]
target_col = 'happiness_score'

X = df[feature_cols]
y = df[target_col]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

imputer = SimpleImputer(strategy='median')
X_train_imp = imputer.fit_transform(X_train)
X_test_imp = imputer.transform(X_test)
y_train = y_train.fillna(y_train.median())
y_test = y_test.fillna(y_test.median())

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train_imp)
X_test_scaled = scaler.transform(X_test_imp)
```

```python
from sklearn.neural_network import MLPRegressor
from sklearn.linear_model import Ridge
from sklearn.metrics import r2_score, mean_squared_error
mlp_model = MLPRegressor(
    hidden_layer_sizes=(100, 50),
    activation='relu',
    solver='adam',
    max_iter=1000,
    random_state=42
)
mlp_model.fit(X_train_scaled, y_train)

ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X_train_scaled, y_train)

y_pred_mlp = mlp_model.predict(X_test_scaled)
y_pred_ridge = ridge_model.predict(X_test_scaled)

r2_mlp = r2_score(y_test, y_pred_mlp)
mse_mlp = mean_squared_error(y_test, y_pred_mlp)

r2_ridge = r2_score(y_test, y_pred_ridge)
mse_ridge = mean_squared_error(y_test, y_pred_ridge)

print(f"MLPRegressor -> R^2: {r2_mlp:.4f}, MSE: {mse_mlp:.4f}")
print(f"Ridge        -> R^2: {r2_ridge:.4f}, MSE: {mse_ridge:.4f}")
```

```python
from matplotlib import pyplot as plt

plt.figure(figsize=(8, 5))
plt.plot(mlp_model.loss_curve_, color='blue', lw=2)
plt.title('Loss Curve моделі MLPRegressor')
plt.xlabel('Ітерації')
plt.ylabel('Loss number')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('mlp_loss_curve.png', dpi=150)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

models_preds = [
    ('MLPRegressor', y_pred_mlp, r2_mlp, 'blue'),
    ('Ridge Regression', y_pred_ridge, r2_ridge, 'green')
]

for ax, (name, y_pred, r2, color) in zip(axes, models_preds):
    ax.scatter(y_test, y_pred, color=color, alpha=0.5, edgecolors='black')
    ax.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
    ax.set_title(f'{name}\nR² = {r2:.4f}')
    ax.set_xlabel('Фактичний Happiness Score')
    if ax == axes[0]:
        ax.set_ylabel('Передбачений Happiness Score')
    ax.grid(alpha=0.3)

plt.tight_layout()
plt.savefig('models_comparison.png', dpi=150)
plt.show()
```
